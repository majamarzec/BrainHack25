```python
import os
import shutil

def move_csv_to_main_folder(main_folder):
    # Przechodzimy przez wszystkie foldery i pliki w katalogu głównym (rekursywnie)
    for root, dirs, files in os.walk(main_folder, topdown=True):  # topdown=True, żeby najpierw sprawdzać foldery
        for file in files:
            if file.endswith('.csv'):
                # Tworzymy pełną ścieżkę pliku
                file_path = os.path.join(root, file)
                
                # Tworzymy ścieżkę docelową w głównym folderze
                destination_path = os.path.join(main_folder, file)
                
                # Sprawdzamy, czy plik już istnieje w głównym folderze
                if not os.path.exists(destination_path):
                    shutil.move(file_path, destination_path)
                    print(f"Przeniesiono: {file_path} -> {destination_path}")
                else:
                    print(f"Plik już istnieje: {file}")

# Wywołanie funkcji
main_folder = 'vertical'
move_csv_to_main_folder(main_folder)
```
