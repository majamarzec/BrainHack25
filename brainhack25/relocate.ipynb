```python
import os
import shutil

def move_csv_to_main_folder(main_folder):
    # Przechodzimy przez wszystkie foldery i pliki w katalogu gÅÃ³wnym (rekursywnie)
    for root, dirs, files in os.walk(main_folder, topdown=True):  # topdown=True, Å¼eby najpierw sprawdzaÄ foldery
        for file in files:
            if file.endswith('.csv'):
                # Tworzymy peÅnÄ ÅcieÅ¼kÄ pliku
                file_path = os.path.join(root, file)
                
                # Tworzymy ÅcieÅ¼kÄ docelowÄ w gÅÃ³wnym folderze
                destination_path = os.path.join(main_folder, file)
                
                # Sprawdzamy, czy plik juÅ¼ istnieje w gÅÃ³wnym folderze
                if not os.path.exists(destination_path):
                    shutil.move(file_path, destination_path)
                    print(f"Przeniesiono: {file_path} -> {destination_path}")
                else:
                    print(f"Plik juÅ¼ istnieje: {file}")

# WywoÅanie funkcji
main_folder = 'vertical'
move_csv_to_main_folder(main_folder)
```
